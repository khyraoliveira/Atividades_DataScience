# Atividade - Script Python: Menu Interativo.

```python
def exibir_menu():
    print('<~~~~~ MENU INICIAL ~~~~~>')
    print('1 - Criar Turma')
    print('2 - Adicionar Professor')
    print('3 - Adicionar Estudante')
    print('4 - Adicionar Nota')
    print('5 - Consultar Nota')
    print('6 - Salvar Nota')
    print('7 - Sair')

while True:
    exibir_menu()
    # usando o 'try' porque o usuário pode digitar algo que NÃO seja um 'número'.
    # caso isso ocorra (o usuário digitar algo inválido): o 'except' é chamado e aparece a msg de erro!
    try:
        opcao = int(input('Escolha uma opção: '))
        
        if opcao == 1:
            print('-- CRIAR TURMA --\nDigite o nome da Turma a ser criada: ')
        elif opcao == 2:
            print('-- ADICIONAR PROFESSOR --\nDigite o nome do Professor a ser adicionado: ')
        elif opcao == 3:
            print('-- ADICIONAR ESTUDANTE --\nDigite o nome do Estudante a ser adicionado: ')
        elif opcao == 4:
            print('-- ADICIONAR NOTA --\nDigite a nota a ser adicionada: ')
        elif opcao == 5:
            print('-- CONSULTAR NOTA --\nDigite a nota que deseja consultar: ')
        elif opcao == 6:
            print('-- SALVAR NOTA --\nDeseja salvar essa nota?')
        elif opcao == 7:
            print('Saindo do Menu...\nAguarde um momento...\nPrograma finalizado com sucesso!')
            break
        # Ao escolher o '7': o break é executado e o loop 'while True' é interrompido, finalizando o programa!
        else:
            print('Opção escolhida inválida. Por favor, tente novamente.')
        # Caso o usuário digite OUTRO número (fora de 1 a 7), a msg de 'tentar novamente' é exibida.
    except ValueError:
        print('O dado digitado é inválido. Insira um número correspondente às opções.')
```
